```python
import tkinter as tk
from tkinter import messagebox

# File to store history
History_File = "history.txt"

# Save a calculation to history
def save_history(equation, result):
    with open(History_File, "a") as file:
        file.write(equation + " = " + str(result) + "\n")

# Show calculation history
def show_file():
    try:
        with open(History_File, "r") as file:
            lines = file.readlines()
            if not lines:
                messagebox.showinfo("History", "No history found!")
            else:
                history = "\n".join(reversed([line.strip() for line in lines]))
                messagebox.showinfo("Calculation History", history)
    except FileNotFoundError:
        messagebox.showinfo("History", "No history file found!")

# Clear the history file
def clear_history():
    with open(History_File, "w") as file:
        pass
    messagebox.showinfo("History", "History cleared successfully!")

# Handle button clicks
def click(event):
    button_text = event.widget.cget("text")
    if button_text == "=":
        try:
            equation = entry.get()
            result = eval(equation)
            entry.delete(0, tk.END)
            entry.insert(tk.END, str(result))
            save_history(equation, result)
        except:
            entry.delete(0, tk.END)
            entry.insert(tk.END, "Error")
    elif button_text == "C":
        entry.delete(0, tk.END)
    else:
        entry.insert(tk.END, button_text)

# Setup window
root = tk.Tk()
root.geometry("330x470")
root.title("Python GUI Calculator with History")

# Entry field
entry = tk.Entry(root, font="Arial 20")
entry.pack(fill=tk.BOTH, ipadx=8, pady=10, padx=10)

# Calculator buttons
buttons = [
    ['7', '8', '9', '/'],
    ['4', '5', '6', '*'],
    ['1', '2', '3', '-'],
    ['C', '0', '=', '+']
]

for row in buttons:
    frame = tk.Frame(root)
    for btn_text in row:
        button = tk.Button(frame, text=btn_text, font="Arial 18", height=2, width=5)
        button.pack(side=tk.LEFT, padx=5, pady=5)
        button.bind("<Button-1>", click)
    frame.pack()

# Add history and clear history buttons
bottom_frame = tk.Frame(root)
history_btn = tk.Button(bottom_frame, text="ðŸ“œ View History", font="Arial 12", command=show_file)
clear_btn = tk.Button(bottom_frame, text="ðŸ—‘ Clear History", font="Arial 12", command=clear_history)
history_btn.pack(side=tk.LEFT, padx=10, pady=10)
clear_btn.pack(side=tk.LEFT, padx=10, pady=10)
bottom_frame.pack()

root.mainloop()
```
